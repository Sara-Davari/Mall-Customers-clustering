```python
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Uploading the dataset
df = pd.read_csv('Mall_Customers.csv')
```

```python
# Viewing the dataset
df.head()
```

```python
# Creating X
X = df[["Age" , "Annual Income (k$)" , "Spending Score (1-100)"]]
```

```python
# Making the data standard
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Creating clustering model and column
kmeans = KMeans(n_clusters = 5)
df["cluster"] = kmeans.fit_predict(X_scaled)
```

```python
# 3d plot to see the clusters
scatterplot_3d = px.scatter_3d(df , x = 'Age' , y = 'Annual Income (k$)' , z = 'Spending Score (1-100)' , color = 'cluster' , color_continuous_scale = 'RdYlGn')
scatterplot_3d.show()
```

```python
# Creating PCA
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Splitting the data into 2 PCAs
df["PCA_1"] = X_pca[ : , 0]
df["PCA_2"] = X_pca[ : , 1]
```

```python
# Viewing the new dataset
df.head()
```

```python
# 2d plot to see the relationship of PCAs and clusters
pca_plot = px.scatter(df , x = 'PCA_1' , y = 'PCA_2', color = 'cluster' , color_continuous_scale = 'jet')
pca_plot.show()
```

```python
# Determining the average of each column in every cluster
cluster_sum = df.groupby("cluster")[["Age" , "Annual Income (k$)" , "Spending Score (1-100)"]].mean()
cluster_sum
```
